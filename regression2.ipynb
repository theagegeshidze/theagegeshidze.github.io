```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import pylab as plt
```

```python
df=pd.read_csv('conclusiveregressiondata.csv')
```

```python
df
```

```python
df2=df.dropna()
```

```python
df2
```

```python
from pandas.io.formats.info import DataFrameTableBuilder
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = df2[['CO2 Emissions per capita']]
y = df2[['Low carbon energy percentage']]
```

```python
regressor = LinearRegression()
regressor.fit(X, y)


y_pred = regressor.predict(X)
# Calculates R squared
score = regressor.score(X,y)
```

```python
from scipy.stats import linregress
```

```python
import numpy as np
```

```python
slope, intercept, rvalue, pvalue, stderr = linregress(x=df2.dropna()['CO2 Emissions per capita'], y=df2.dropna()['Low carbon energy percentage'])
```

```python
print('y = ',np.round(slope,2),' * x + ',np.round(intercept,2))
print('R = ',np.round(rvalue**2,2))
```

```python
plt.annotate(f'R\u00b2 = {score:.3f}', xy=(0.5,0.7), xycoords='axes fraction')
plt.scatter(X, y, color= 'pink')
plt.plot(X, regressor.predict(X), color='tab:pink')
plt.title('Low Carbon Energy on CO2 Emissions', fontsize=15)
plt.xlabel('CO2 Emissions per capita', fontsize=10)
plt.ylabel('Low carbon energy percentage', fontsize=10)
plt.savefig('firstregression')
plt.show()
```
