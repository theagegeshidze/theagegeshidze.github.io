```python
import requests
import json
from google.colab import files

# imported relevant packages

url_base = "https://api.ons.gov.uk/timeseries/{}/dataset/PUSF/data"
# insert the url value of the office of national statistics api
file_base = "data_ONS-{}.json"
# this is the file base form of the loop
onsSeries = ['NMBY', 'NZGF', 'MF6V', 'GTAO', 'CUDG', 'MM9F', 'CUKY', 'BKST', 'EKED']
# these are the series values for the data I want to loop, in this case I have chosen tax revenues, so I can create a tax dashboard

for i in onsSeries:  
   
    print("------Iteration Starts--------")
    print(i) 
  

    URL = url_base.format(i)
    print(URL)
   
    data = requests.get(URL).json()
    print(data)


    fileName = file_base.format(i)
    print(fileName)
   
    print("------Iteration Ends--------")
   

    with open(fileName, 'w', encoding='utf-8') as f:
     json.dump(data, f, ensure_ascii=False, indent=4)
    files.download(fileName)

    # i have printed the iteration and looped it and now I get the appropriate data for my json
    # i can now use the data in my vega json file, and simpy swap out the codes 9 times and so easilly
    # create 9 graphs for my dashboard!!!

    # was really proud of myself with this one as I found it really complicated to understand the concept of APIs at first
```
